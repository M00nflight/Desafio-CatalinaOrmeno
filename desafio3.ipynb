```python
# ============================================================
# DESAFÍO 3
# Autor: Catalina Ormeño Rivera
# Fecha: 23/10/2025
# Archivo: desafio3.ipynb
# ============================================================

# Importar Librería necesaria
import pandas as pd
# Cargar archivo original de transacciones (transacciones.txt)
df = pd.read_csv("transacciones.txt", sep="|")

# Mostrar las primeras filas para ver estructura
print("Datos originales:")
display(df.head())


# Separar la columna compuesta 'canal_venta'
# Esta columna tiene dos valores separados por espacio (canal + sucursal)
df[['canal', 'sucursal']] = df['canal_venta'].str.split(' ', expand=True)

# 4Reemplazar los códigos por nombres legibles
df['canal'] = df['canal'].map({'CtoCcCl': 'Call Center Cliente'})
df['sucursal'] = df['sucursal'].map({'CTOMA': 'Sucursal Toma'})


# Eliminar columna original y limpiar datos
df = df.drop(columns=['canal_venta'])

# Mostrar resultado limpio
print("Datos transformados:")
display(df.head())

# Exportar el dataframe limpio a CSV
df.to_csv("transacciones_clean.csv", index=False)
print("Archivo limpio exportado como 'transacciones_clean.csv'")

# Demostración final
print(f"Total de registros procesados: {len(df)}")
print(f"Columnas finales: {list(df.columns)}")
```
